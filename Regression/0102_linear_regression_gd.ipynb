```python
import numpy as np
import matplotlib.pyplot as plt
```

## 데이터 준비

```python
X_train = np.array([1., 2., 3., 4., 5., 6.])
y_train = np.array([9., 12., 15., 18., 21., 24.])

# y = 3x + 6
```

```python
plt.scatter(X_train, y_train)
```

## 학습
- 경사하강법 (Gradient Descent)

```python
W = 0.0
b = 0.0
N = len(X_train)

epochs = 5000
lr = 0.01

for i in range(epochs):
  # 가설
  h_hat = W * X_train + b

  # Cost. MSE
  cost = np.sum( (y_train - h_hat) ** 2 ) / N

  # 미분
  grad_W = np.sum((W * X_train + b - y_train) * 2 * X_train) / N
  grad_b = np.sum((W * X_train + b - y_train) * 2) / N
  
  W = W - lr * grad_W
  b = b - lr * grad_b

  if i % 200 == 0:
    print(f'Epoch:{i}, cost:{cost:10f}, W:{W:10f}, b:{b:10f}')
```

a=3, b=6을 맞춘다

즉, h_hat = 3.0 * X_train + 6.0

## 클래스

```python
class LinearRegressor:
  def fit(self, X, y, epochs=100, lr=0.01):
    self.W_ = 0.0
    self.b_ = 0.0
    self.N_ = len(X)
    self.lr_ = lr

    for i in range(epochs):
      h_hat = self.W_ * X+ self.b_

      cost = np.sum( (y - h_hat) ** 2 ) / self.N_

      grad_W = np.sum((self.W_ * X + self.b_ - y) * 2 * X) / self.N_
      grad_b = np.sum((self.W_ * X + self.b_ - y) * 2) / self.N_
  
      self.W_ = self.W_ - self.lr_ * grad_W
      self.b_ = self.b_ - self.lr_ * grad_b

      if i % 200 == 0:
        print(f'Epoch:{i}, cost:{cost:10f}, W:{self.W_:10f}, b:{self.b_:10f}')

  def predict(self, X):
    return self.W_ * X + self.b_
```

```python
lr = LinearRegressor()
lr.fit(X_train, y_train, epochs = 5000)
```

```python
X_test = np.array([10, 11])
y_pred = lr.predict(X_test)
y_pred
```
