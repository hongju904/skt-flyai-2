```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

## 데이터 준비

```python
X_train = np.array([1., 2., 3., 4., 5., 6.])
y_train = np.array([9., 12., 15., 18., 21., 24.])
```

## 모델

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(units=1, input_shape=(1,)), # 시퀀스 안에 레이어를 넣고(Dense), 그 안에 유닛을 넣음
    # 첫번째 레이어에서 input_shape를 tensor 형태로 작성
    # layers.Dense(units=1,), # 레이어를 하나 더 생성
])
```

```python
model.summary()
```

Dense가 하나 -> 인풋은 상관없고, 출력(유닛)은 하나

학습시킬 파라미터는 w, b 2개

```python
# 컴파일
adam = tf.keras.optimizers.Adam(learning_rate=0.01)
sgd = tf.keras.optimizers.experimental.SGD()

model.compile(
    optimizer = sgd, # 어떤 GD를 쓸 것인가
    loss = 'mse', # loss 함수
    metrics = ['mse', 'mae'] # 평가지표
)
```

## 학습

```python
model.fit(X_train, y_train, epochs=100, verbose=1)
```

```python
# 학습된 파라미터 확인
model.get_weights()
```

## 평가

```python
model.evaluate(X_train, y_train)
```
