# 선형회귀 (Gradient Tape)

- 텐서플로우의 자동미분 기능

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

## 데이터 준비

```python
X_train = np.array([1., 2., 3., 4., 5., 6.])
y_train = np.array([9., 12., 15., 18., 21., 24.])

# y = 3x + 6
```

## 경사하강법

```python
# 초기화
W = tf.Variable(0.0)
b = tf.Variable(0.0)
  # np.array로 바꾸고 싶을 경우 -> W.numpy()
y_hat = W * X_train + b
```

```python
# MSE :: 제곱 후 평균 구하기
cost = tf.reduce_mean( tf.square(y_train - y_hat) )
```

```python
# 자동미분 기능
lr = 0.01

with tf.GradientTape() as tape:
  y_hat = W * X_train + b
  cost = tf.reduce_mean( tf.square(y_train - y_hat) )
  # 미분값만 저장 (아직 대입 X)

W_grad, b_grad = tape.gradient(cost, [W, b])

W.assign_sub(lr * W_grad)
  # 빼기를 해서 덮어씌움
  # W = W - lr * W.grad 와 동일
b.assign_sub(lr * b_grad)
```

```python
W.numpy(), b.numpy()
```

## 전체 코드

```python
# 초기화
W = tf.Variable(0.0)
b = tf.Variable(0.0)
epochs = 5000
lr = 0.01


for i in range(epochs):
  with tf.GradientTape() as tape:
    y_hat = W * X_train + b
    cost = tf.reduce_mean( tf.square(y_train - y_hat) )

    W_grad, b_grad = tape.gradient(cost, [W, b])

    W.assign_sub(lr * W_grad)
    b.assign_sub(lr * b_grad)

    if i % 200 == 0:
      print(f'Epoch:{i}, cost:{cost.numpy():10f}, W:{W.numpy():10f}, b:{b.numpy():10f}')
```
