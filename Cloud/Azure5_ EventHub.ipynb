```python
import time
from datetime import datetime

while(1):
  message = 'time: '+str(datetime.now())
  message = EventData(message) # 보낼 수 있는 형태로 변환
  eventhub.send_event(message) # 1건마다 보낸다

  time.sleep(2)                # 2초 단위 전송
```

```python
# 데이터를 가져오는 코드 (원래 수신코드는 다른 파일에 있어야 함)
import asyncio
from azure.eventhub.aio import EventHubConsumerClient

async def on_event(partition_context, event):
  print('Received the event: \"{}\" from the partition with ID: \"{}\"'
        .format(event.body_as_str(encoding='UTF-8')), partition_context)
  
  print(event)
  await partition_context.update_checkpoint(event)

client = EventHubConsumerClient.from_connection_string("Endpoint=sb://labuser20eventhubs.servicebus.windows.net/;SharedAccessKeyName=admin;SharedAccessKey=ZwotiZVyfRALf5zMo59yRn3OFOASIA2YkLZ3TjdQ/9A=;EntityPath=eventhub1", '$Default')
```

```python
client.receive(on_event=on_event)
```
